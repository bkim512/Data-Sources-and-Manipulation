Using the first citation above, I replaced the incorrect name of the village by looking at the clean data.

```python
data['village'].replace('49', 'Chirodzo', inplace=True)
data['village'].replace('Ruaca-Nhamuenda', 'Ruaca', inplace=True)
data['village'].replace('Ruaca - Nhamuenda', 'Ruaca', inplace=True)
data['village'].replace('Ruca', 'Ruaca', inplace=True)
data['village'].replace('Chirdozo', 'Chirodzo', inplace=True)
```

Challenge Question:
The value '49' was supposed to be Chirodzo. I used SAFI_clean.csv file that was found in the link of information about that data (Data Codebook). Everything was in the same order so I looked at the village of line 50 which was "Chirodzo"

```python
#months_no_water and months_lack_food for village God
#without NaN
village_god = data.loc[data['village'] == 'God', ['months_no_water','months_lack_food' ]]
no_god = village_god.dropna()
no_god
```

```python
#the number of months and which months the families interviewed do not have enough food or water for village God
no_god = no_god.copy()
god_count_no_water = []
god_count_no_food = []
for index,row in no_god.iterrows():
    god_count_no_water.append(len(row['months_no_water'].split(';')))
    god_count_no_food.append(len(row['months_lack_food'].split(';')))
no_god['god_count_no_water'] = god_count_no_water
no_god['god_count_no_food'] = god_count_no_food
no_god
```

```python
#to save my village God data
no_god.to_csv('village_god_data.csv')
```
