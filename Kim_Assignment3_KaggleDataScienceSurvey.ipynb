Then, I have counted how many times each language was recommeneded to compare which one was most recommended

```python
Python = 0
SQL = 0
R = 0
Other = 0
Scala = 0
None_language = 0
Java = 0
MATLAB = 0
C_plusplus = 0
SAS = 0
Javascript = 0
Go = 0
VBA = 0
for value in new_data['Q18']:
    if value == 'Python':
        Python += 1
    elif value == 'SQL':
        SQL += 1
    elif value == 'R':
        R += 1
    elif value == 'Other':
        Other += 1
    elif value == 'Scala':
        Scala += 1
    elif value == 'None':
        None_language += 1
    elif value == 'Java':
        Java += 1
    elif value == 'MATLAB':
        MATLAB += 1
    elif value == 'C++':
        C_plusplus += 1
    elif value == 'SAS':
        SAS += 1
    elif value == 'Javascript':
        Javascript += 1
    elif value == 'Go':
        Go += 1
    elif value == 'VBA':
        VBA += 1
```

From the list above, it looks like Python has highest number of recommendation. However, I made a pie chart and a bar graph to visually compare recommendation of each programming languages

```python
list_of_labels=['Python', 'SQL', 'R', 'Other', 'Scala', 'None', 'Java', 'MATLAB',
       'C++', 'SAS', 'Javascript', 'Go', 'VBA']
```

```python
plt.pie(list_of_values, labels=list_of_labels, autopct='%1.1f%%', startangle=90)
```

From the pie chart and bar graph above, it looks like top three recommendation of programming languages are Python with 75.5% of the votes, R with 12.5% of the votes, and SQL with 4.8% of the votes.
