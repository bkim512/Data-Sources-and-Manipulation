### Q8) Calculating Values (4pts)
Create a new column named `gdpPercap_multiple` in the dataframes for european and american regions. In this column save the 2007 GDP per capita for each country divided by GDP per capita for that country in 1952 with only 2 decimal places of precision.
```python
#creating new column for europe_df
europe_df['gdpPercap_multiple'] = europe_df['gdpPercap_2007']/europe_df['gdpPercap_1952']
#only 2 decimal places
europe_df = europe_df.round({'gdpPercap_multiple': 2})
#creating new column for americas_df
americas_df['gdpPercap_multiple'] = americas_df['gdpPercap_2007']/americas_df['gdpPercap_1952']
#only 2 decimal places
americas_df = americas_df.round({'gdpPercap_multiple': 2})
```
