Raw-numbers don't often tell the story in the most effective manner. We will calculate the percentage for the demographics we are interested in.

**(18)** Create a columns called 'female_percent' and 'male_percent' that contains the percentage of the total number that voted that identified by the respective gender.

```python
# FILL-IN 18
num_men = pd.Series(votes['Male'])
num_women = pd.Series(votes['Female'])
num_total = pd.Series(votes['total_voted'])
int_men = pd.to_numeric(num_men)
int_women = pd.to_numeric(num_women)
int_total = pd.to_numeric(num_total)
votes['male_percent'] = (int_men/int_total)*100
votes['female_percent'] = (int_women/int_total)*100
votes
```

__(19)__ Create columns for each age group that contains the percent of people in that group that voted out of the number registered in that group (for example, the percentage of 18-24 year-olds that voted out of the number of 18-24 year-olds that registered).

```python
# FILL-IN 19
v_1824 = pd.Series(votes['18-24 Voted'])
r_1824 = pd.Series(votes['18-24 Registered'])
int_v_1824 = pd.to_numeric(v_1824)
int_r_1824 = pd.to_numeric(r_1824)
votes['18-24 Voted Percent'] = (int_v_1824/int_r_1824)*100

v_2544 = pd.Series(votes['25-44 Voted'])
r_2544 = pd.Series(votes['25-44 Registered'])
int_v_2544 = pd.to_numeric(v_2544)
int_r_2544 = pd.to_numeric(r_2544)
votes['25-44 Voted Percent'] = (int_v_2544/int_r_2544)*100

v_4564 = pd.Series(votes['45-64 Voted'])
r_4564 = pd.Series(votes['45-64 Registered'])
int_v_4564 = pd.to_numeric(v_4564)
int_r_4564 = pd.to_numeric(r_4564)
votes['45-64 Voted Percent'] = (int_v_4564/int_r_4564)*100

v_65up = pd.Series(votes['65 and over Voted'])
r_65up = pd.Series(votes['65 and over Registered'])
int_v_65up = pd.to_numeric(v_65up)
int_r_65up = pd.to_numeric(r_65up)
votes['65 and over Voted Percent'] = (int_v_65up/int_r_65up)*100

votes
```

Create a stacked bar chart from this demo: https://matplotlib.org/gallery/lines_bars_and_markers/bar_stacked.html#sphx-glr-gallery-lines-bars-and-markers-bar-stacked-py
